###Maternal Mental Health and Infant Sleep Analysis

Project objective

Maternal Mental Health and Infant Sleep Analysis
The objective of this project is to explore the relationship between maternal mental health and infant sleep characteristics using a dataset containing demographic, psychological, and infant sleep variables.
Specifically, the analysis aims to:
Examine maternal depression using EPDS scores.
Examine maternal anxiety using HADS scores.
Examine childbirth-related trauma using CBTS scores.
Investigate the relationship between maternal mental health and infant sleep patterns.
Analyze the effect of demographic factors such as age, education, marital status, and baby sex on maternal mental health outcomes.
Identify important factors associated with maternal well-being during early motherhood.
